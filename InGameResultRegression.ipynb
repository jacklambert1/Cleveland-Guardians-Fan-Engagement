```python
# Import necessary libraries

import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# Read in data

stats_twitter = pd.read_csv('/content/drive/MyDrive/IndependentProjects/GuardiansProject/CSVs/monthStatsAndTwitter.csv')
```

```python
# Create linear regression based on "excitement measures": HR, runs scores, hits, strikeOutsPitching on per game basis

# Load data
stats_twitter1 = pd.DataFrame(stats_twitter)
new_stats_twitter = stats_twitter1.dropna() # Drop na

# Create per game columns
new_stats_twitter['HR/Game'] = new_stats_twitter['homeRuns']/new_stats_twitter['totalGames']
new_stats_twitter['RunsScored/Game'] = new_stats_twitter['runsScored']/new_stats_twitter['totalGames']
new_stats_twitter['Hits/Game'] = new_stats_twitter['hits']/new_stats_twitter['totalGames']
new_stats_twitter['StrikeOuts/Game'] = new_stats_twitter['strikeOutsPitching']/new_stats_twitter['totalGames']

# Separate data into predictors and prediction value
X = new_stats_twitter[['HR/Game', 'RunsScored/Game', 'Hits/Game', 'StrikeOuts/Game']]
Y = new_stats_twitter['followerDiff']

# Create regression
regr = linear_model.LinearRegression()
regr.fit(X, Y)
predict = regr.predict(X)

# Print regression stats
print(f'Coefficients: {regr.coef_}')
print(f'Intercept: {regr.intercept_}')
print(f'R^2 Value: {r2_score(Y,predict)}\n\n\n\n')
```
