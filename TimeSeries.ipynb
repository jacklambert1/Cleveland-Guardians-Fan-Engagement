```python
# Import necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
guardians_engagement = pd.read_csv('/content/drive/MyDrive/IndependentProjects/GuardiansProject/CSVs/guardians_engagement.csv')
```

```python
# Set engagement data to be only for specific player selected (Jose Ramirez)
PE = guardians_engagement[guardians_engagement['playerId']==608070]
PE.index = PE['engagementMetricsDate']
```

```python
# All 4 metrics plot together
PE['target1'].plot(linewidth=2, figsize=(15,10),legend=True, color='purple')
PE['target2'].plot(linewidth=2, figsize=(15,10),legend=True, color='green')
PE['target3'].plot(linewidth=2, figsize=(15,10),legend=True, color='orange')
PE['target4'].plot(linewidth=2, figsize=(15,10),legend=True, color='blue')
```

```python
# target1 plot
PE['target1'].plot(linewidth=2, figsize=(15,10),legend=True, color='purple')
PE['1MA25'] = PE['target1'].rolling(25).mean()
PE['1MA25'].plot(linewidth=2, figsize=(15,10),legend=True, color='red')
```

```python
# target2 plot
PE['target2'].plot(linewidth=2, figsize=(15,10),legend=True, color='green')
PE['2MA25'] = PE['target2'].rolling(25).mean()
PE['2MA25'].plot(linewidth=2, figsize=(15,10),legend=True, color='red')
```

```python
# target3 plot 
PE['target3'].plot(linewidth=2, figsize=(15,10),legend=True, color='orange')
PE['3MA25'] = PE['target3'].rolling(25).mean()
PE['3MA25'].plot(linewidth=2, figsize=(15,10),legend=True, color='red')
```

```python
# target4 plot 
PE['target4'].plot(linewidth=2, figsize=(15,10),legend=True, color='blue')
PE['4MA25'] = PE['target4'].rolling(25).mean()
PE['4MA25'].plot(linewidth=2, figsize=(15,10),legend=True, color='red')
```

```python
# Create nicely formatted plot of target 2 to put in final presentation

# colors
navyBlue = '#073763'
red = '#ff0000ff'
white = '#FFFFFF'

# Initialize figure and axis
fig, ax = plt.subplots()
ax.set_facecolor(white)

# Plot values and moving average
ax.plot(PE.index, PE['target2'], color = navyBlue, label = 'Observed Value')
ax.plot(PE.index, PE['2MA25'], color = red, label = 'Moving Average')

# Set labels
ax.set_title('Player Engagement Over Time', color = navyBlue)
ax.set_xlabel('Day', color = navyBlue)
ax.set_ylabel('Engagement Target 2', color = navyBlue)

# create tick marks
ax.set_xticks(ticks = [])
ax.set_yticklabels(labels = np.arange(0,81,10), color = navyBlue)

ax.tick_params(length = 0)

ax.spines['bottom'].set_color(navyBlue) #include x-axis line
ax.spines['top'].set_color(navyBlue)
ax.spines['left'].set_color(navyBlue)
ax.spines['right'].set_color(navyBlue)

# Create legend
l = plt.legend()
l.get_frame().set_edgecolor(navyBlue)
for text in l.get_texts():
    text.set_color(navyBlue)

plt.show()
```
